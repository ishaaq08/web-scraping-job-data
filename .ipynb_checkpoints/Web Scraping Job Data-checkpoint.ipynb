```python
from bs4 import BeautifulSoup
import lxml.etree as ET
import requests
import random
```

```python
# Optimizing the request headers

HEADERS = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:98.0) Gecko/20100101 Firefox/98.0",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
        "Accept-Language": "en-US,en;q=0.5",
        "Accept-Encoding": "gzip, deflate",
        "Connection": "keep-alive",
        "Upgrade-Insecure-Requests": "1",
        "Sec-Fetch-Dest": "document",
        "Sec-Fetch-Mode": "navigate",
        "Sec-Fetch-Site": "none",
        "Sec-Fetch-User": "?1",
        "Cache-Control": "max-age=0",
    }

url = "https://www.gradcracker.com/search/computing-technology/graduate-jobs?order=dateAdded"

page = requests.get(url, headers=HEADERS)
soup = BeautifulSoup(page.text, "lxml")
```

```python
# class="tw-relative tw-mb-4 tw-border-2 tw-border-gray-200 tw-rounded"

first_item = soup.find("div", class_ = "tw-relative tw-mb-4 tw-border-2 tw-border-gray-200 tw-rounded")
```

```python
first_item
```

```python
title = first_item.find("a", class_ = "tw-block tw-text-base tw-font-semibold")
title_final = title.text.strip()
```

```python
# Industry - tw-text-xs tw-font-bold tw-text-gray-800

industry = first_item.find("div", class_ = "tw-text-xs tw-font-bold tw-text-gray-800")
industry_final = industry.text.strip()
```

```python
# Salary

salary = first_item.find_all("li")[0].text.split()[-1]
salary
```

```python
# Location

location_unsplit = first_item.find_all("li")[1].text.strip()
location_split = location_unsplit.split(":")[-1]
location_final = location_split.strip()
```

```python
# Company

title
```

```python
company = title.attrs["title"].split("with")[-1].strip()
```

```python
import pandas as pd 

df = pd.DataFrame(columns=["Title", "Company", "Industry", "Salary", "Location"])
df
```

```python
df.loc[0] = [title_final, company, industry_final, salary,  location_final]
```

```python
df
```

```python
# Actual Script

job_postings = soup.find_all("div", class_ = "tw-relative tw-mb-4 tw-border-2 tw-border-gray-200 tw-rounded")
counter = 1
for job in job_postings:
    job_title = job.find("a", class_ = "tw-block tw-text-base tw-font-semibold").text.strip()
    job_company = job.find("a", class_ = "tw-block tw-text-base tw-font-semibold").attrs["title"].split("with")[-1].strip()
    industry_temp = job.find("div", class_ = "tw-text-xs tw-font-bold tw-text-gray-800")
    if industry_temp == None:
        job_industry = None
    else:
        job_industry = industry_temp.text
    job_salary = job.find_all("li")[0].text.split(": ")[-1]
    job_location = job.find_all("li")[1].text.strip().split(": ")[-1]
    item_to_add = [job_title, job_company, job_industry, job_salary, job_location]
    df.loc[len(df)] = item_to_add
    
    
```
